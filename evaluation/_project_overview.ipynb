# O-RAG Project: Complete Evaluation Framework

## 📖 Welcome to the O-RAG Evaluation Suite

This is a **comprehensive research evaluation** of the **Offline Retrieval-Augmented Generation (O-RAG)** system.

### What is O-RAG?
O-RAG is a system that combines:
- **Retrieval**: Finding relevant documents/chunks from a knowledge base
- **Generation**: Using an LLM to generate answers based on retrieved context
- **Offline**: Runs entirely locally without cloud dependencies
- **Multi-strategy**: Tries multiple retrieval approaches and combines them

### 🎯 Project Goal
Compare **7 different retrieval strategies** across **3 domains** (Healthcare, Technical, Financial) to find:
- Which strategy works best overall?
- Does the best strategy change by domain?
- Is there a speed vs. quality trade-off?
- What does an independent LLM judge think about the quality?

### 📊 Evaluation Scope
- **252 total evaluations** (7 variants × 3 domains × 12 questions)
- **6 retrieval metrics**: Recall, Precision, Hit Rate, MRR, NDCG@5, and latency
- **4 LLM quality metrics**: Faithfulness, Relevance, Coherence, Completeness
- **Multiple comparisons**: Domain-specific, variant-specific, and cross-cutting

## 📋 What Gets Measured?

### Retrieval Quality Metrics

**Context Recall** (0-1)
- Did we retrieve the relevant information?
- Questions: "Did you find the important keywords/facts?"
- Best: 1.0 (all relevant info found)

**Context Precision** (0-1)
- How much of what we retrieved was actually relevant?
- Questions: "Was everything you retrieved useful?"
- Best: 1.0 (everything retrieved was relevant)

**Hit Rate** (0 or 1)
- Did we find ANY relevant information?
- Questions: "Did you find at least one relevant document?"
- Best: 1 (always found something)

**Mean Reciprocal Rank (MRR)** (0-1)
- How early in the results did we find the first relevant item?
- Questions: "On average, how many items did we need to check?"
- Best: 1.0 (relevant item was first)

**NDCG@5** (0-1)
- Ranking quality: are the best items at the top?
- Questions: "Is your ranking sensible?"
- Best: 1.0 (perfect ranking in top-5)

**E2E Latency** (milliseconds)
- How fast is the entire retrieval process?
- Questions: "How long does it take?"
- Best: < 30ms (fast enough for real-time use)

### LLM Quality Metrics (0-10 scale)

**Faithfulness**
- Is the answer grounded in the retrieved context?
- Detects: Hallucinations, making things up, contradictions

**Relevance**
- Does the answer actually address the question?
- Detects: Off-topic answers, incomplete answers

**Coherence**
- Is the answer logically structured and clear?
- Detects: Rambling, contradictions, confusing explanations

**Completeness**
- Does the answer cover the key aspects of the question?
- Detects: Partial answers, missing important details

### Test Set per Domain
- **12 questions** per domain
- Questions written to require document retrieval
- Ground truth answers included for evaluation
- Keywords tagged for recall/precision measurement

- 7 retrieval variants tested
- = **252 total evaluations**
- Each evaluation scored on 10+ metrics
