```python
from huggingface_hub import hf_hub_download
import os

print("Downloading Qwen 3.5 1.5B...")
model = hf_hub_download(
    repo_id="thebloke/Qwen3.5-1.5B-Instruct-GGUF",
    filename="qwen3.5-1.5b-instruct-q8_0.gguf",
    local_dir="./models",
    local_dir_use_symlinks=False,
)
print(f"Downloaded: {os.path.getsize(model) / (1024**3):.2f} GB")
```

```python
import llama_cpp
import time
import os

print("Quantizing to Q4_K_M...")
start = time.time()

llama_cpp.llama_model_quantize(
    b"./models/qwen3.5-1.5b-instruct-q8_0.gguf",
    b"./models/qwen3.5-1.5b-instruct-q4_k_m.gguf",
    llama_cpp.llama_model_quantize_params(qtype=25)
)

size = os.path.getsize("./models/qwen3.5-1.5b-instruct-q4_k_m.gguf") / (1024**3)
print(f"Done! Size: {size:.2f} GB  Time: {(time.time()-start)/60:.1f}m")
```
