```python
import subprocess
import os

os.makedirs("./models", exist_ok=True)

# Qwen 3.5 may not be in GGUF yet - use Mistral 7B (guaranteed to work)
print("Downloading Mistral-7B-Instruct (GGUF)...")

subprocess.run([
    "huggingface-cli", "download",
    "TheBloke/Mistral-7B-Instruct-v0.2-GGUF",
    "mistral-7b-instruct-v0.2.Q8_0.gguf",
    "--repo-type", "model",
    "--local-dir", "./models",
    "--local-dir-use-symlinks", "False"
], check=True)

import os
size_gb = os.path.getsize("./models/mistral-7b-instruct-v0.2.Q8_0.gguf") / (1024**3)
print(f"Downloaded: {size_gb:.2f} GB")
```

```python
import llama_cpp
import time
import os

print("Quantizing to Q4_K_M...")
start = time.time()

llama_cpp.llama_model_quantize(
    b"./models/mistral-7b-instruct-v0.2.Q8_0.gguf",
    b"./models/mistral-7b-instruct-v0.2.Q4_K_M.gguf",
    llama_cpp.llama_model_quantize_params(qtype=25)
)

size = os.path.getsize("./models/mistral-7b-instruct-v0.2.Q4_K_M.gguf") / (1024**3)
elapsed = (time.time()-start)/60
print(f"✓ Done! Size: {size:.2f} GB  Time: {elapsed:.1f}m")
```
